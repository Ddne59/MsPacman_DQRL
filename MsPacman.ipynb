```python
env = gym.make("ALE/MsPacman-v5", render_mode="human")

model = stable_baselines3.DQN.load("dqn_mspacman")

# Test the model
obs, info = env.reset()
while True:
    action, _states = model.predict(obs, deterministic=True)
    obs, rewards, dones, info = env.step(action)[0], env.step(action)[1], env.step(action)[2], env.step(action)[3]
    env.render()
    if dones:
        obs = env.reset()
        print("Episode terminé")
        print(rewards)
        break
```
